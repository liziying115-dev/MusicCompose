## TECHIN 509: Melody Representation

I think the best way to store a sequence of notes like C D E F G A B in Python is by using a **list** as the list keeps everything in order. I also want include how long each note lasts and which octave it’s in, so I can store each note as a tuple. 

```python
melody = [("C4", 1.0), ("D4", 1.0), ("E4", 1.0), ("F4", 1.0), ("G4", 2.0)]
```

The first element is the pitch (like "C4").
The second element is the duration in beats (like 1.0 for a quarter note)

If a melody is like a sentence, then each note is like a word.
A list keeps these “words” in the **correct order**.  

Therefore:
Word -> Note
Sentence -> Melody
Paragraph (multiple sentences) -> Song (multiple melodies combine together). 

For mutiple melodies: 

```python
melodies = [
    [("C4", 1.0), ("D4", 1.0), ("E4", 1.0)],
    [("E4", 0.5), ("B4", 0.5), ("A4", 2.0)]
]
```

Combining several melodies into one long sequence 

```python
# flatten them using a loop
all_notes = []
for melody in melodies:
    all_notes.extend(melody)
```
